## 여러 파일을 한번에 로드하는 방법

```python
file_csv = glob("data/store/*.csv")    #glob("data/store/*_*_*.csv")
file_csv
```

```python
pd.read_csv(file_csv[0])
```

```python
file_list = []
for file_csv_name in file_csv:
    print(file_csv_name)
    temp = pd.read_csv(file_csv_name, low_memory=False)
    file_list.append(temp)
```

```python
len(file_list)
```

```python
df = pd.concat(file_list)
df.shape
```

* 17개의 file을 각각의 data frame으로 읽음 -> list안에 넣음 -> concat으로 합침

* pd.concat() > column 명이 같다면 하나의 data frame으로 합침
